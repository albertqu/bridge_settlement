```python
def pca_proc(ns):
    X = cv2.imread(ns, 0)
    X_m = X - np.mean(X, axis=0)[np.newaxis,:]
    m = X_m.shape[0]
    data_mat = X_m.T @ X_m / m
    u, s, vh = np.linalg.svd(X_m / np.sqrt(m))
    sigma_max = np.max(s)
    ind = np.argmax(s)
    print("Maximum Singular Value is {0}, and the variance is {1}".format(sigma_max, (sigma_max ** 2)))
    x_mean = np.mean(X, axis=0)
    x_mean_norm = x_mean / np.linalg.norm(x_mean)
    var_mean = x_mean_norm.T @ data_mat @ x_mean_norm
    print("Maximum Variance with mean: {0}".format(var_mean))
    pca = PCA(n_components=2)
    pca.fit(X)
    print("Total Variance Explained: ", sum(pca.explained_variance_))
    print("Explained Variance Ratio by the first two components: {0}".format(sum(pca.explained_variance_ratio_)))
    a1, a2 = pca.components_
    print(X.shape, a1.shape)
    e1 = X @ a1
    e2 = X @ a2
    fig = plt.figure(figsize=(20, 20))
    ax = fig.add_subplot(311)
    cb = np.floor(256 * np.array(range(len(e1))) / len(e1))
    ax.scatter(e1, e2, c=cb)
    ax2 = fig.add_subplot(312)
    ax2.scatter(np.arange(len(e1)), cb, c=cb)
    ax3 = fig.add_subplot(313)
    ax3.plot(e1, color='Blue')
```

```python
pca_proc(skewed)
```
